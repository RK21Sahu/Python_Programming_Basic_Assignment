```python
Question 1:

Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

Ans - 
class GenDiv:
    def __init__(self,n):
        self.n = n
        
    def div7(self):
        for i in range(self.n):
            if i%7==0:
                yield i
            i+=1
        
a = GenDiv(50)
b = a.div7()

for j in b:
    print(j)
```

```python
Question 2:
Write a program to compute the frequency of the words from the input. The output should output after sorting the key 
alphanumerically. 

Suppose the following input is supplied to the program:

New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.

Then, the output should be:

2:2

3.:1

3?:1

New:1

Python:5

Read:1

and:1

between:1

choosing:1

or:2

to:1

Ans - 
    
class wordfreq:
    def __init__(self,txt):
        self.txt = txt
        
    def clafreq(self):
        lst = self.txt.split()
        uniq = set(lst)
        
        for i in uniq:
            print(i,":",lst.count(i)) 
            

obj = wordfreq("New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.")        
obj.clafreq()
```

```python
Question 3:

Define a class Person and its two child classes: Male and Female. All classes have a method "getGender" which can print 
"Male" for Male class and "Female" for Female class.

Ans - 

class person:
    def getGender(self):
        print("I am a Person")

class male(person):
    def getGender(self):
        print("I am Male")
        

class female(person):
    def getGender(self):
        print("I am Female")


obj = person()
obj.getGender()

obj = male()
obj.getGender()

obj = female()
obj.getGender()
```

```python
Question 4:
Please write a program to generate all sentences where subject is in ["I", "You"] and verb is in ["Play", "Love"] and the
object is in ["Hockey","Football"].

Ans - 
class makesentence:
    def __init__(self,subject,verb,objects):
        self.subject = subject
        self.verb = verb
        self.objects = objects
        
    def sentence(self):
        for i in self.subject:
            for j in self.verb:
                for k in self.objects:
                    print(i, " ", j, " ", k)
                    
subject = ["I", "You"]
verb = ["Play", "Love"]
objects = ["Hockey","Football"]

obj = makesentence(subject,verb,objects)
obj.sentence()
```

```python
Question 6:
Please write a binary search function which searches an item in a sorted list. The function should return the index of 
element to be searched in the list.

ans - 

def binary_search(lst, item):
    low = 0
    high = len(lst) - 1
    mid = 0
 
    while low <= high:
 
        mid = (high + low) // 2
 
        if lst[mid] < item:
            low = mid + 1
 
        elif lst[mid] > item:
            high = mid - 1
 
        else:
            return mid
 
    return -1


lst = [1,2,3,5,6,9,11,22,66]

binary_search(lst,6)
```
